```python
# Compile time / method overloading polymorphism
class BCA_1st_Semester:
    def student(self, a, b=0, c=0, d=0):
        return a + b + c + d

BCA = BCA_1st_Semester()
print(BCA.student(5))
print(BCA.student(10,7,8))
print(BCA.student(-10,4,9,-5))
print(BCA.student(99,0,4,-12))
```

```python
# Run time / Methode overriding\
class Animal :
  def sound(self):
    print("Animal make a sound")

class Cat(Animal) :
  def sound(self):
    print("meow meow meow")

class Snake(Animal) :
  def sound(self):
    print("Buuussssssssssss")

class cow(Animal) :
  def sound(self):
    print("Ambaaaaaaaa")

obj1 = Animal()
obj2 = Cat()
obj3 = Snake()
obj4 = cow()

obj2.sound()
obj1.sound()
obj3.sound()
obj4.sound()
```
